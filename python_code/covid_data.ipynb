```python
import opendatasets as od
import pandas as pd
import numpy as np

# going to make our full data into a function
def full_data():
    dat=od.download("https://www.kaggle.com/pavan9065/coronavirus-covid19-vaccinations")
    df = pd.read_csv('coronavirus-covid19-vaccinations/vaccine-data.csv', engine = 'python', sep='\s*,\s*')
    return df
#df # full data set
df = full_data();
df
```

```python
df_usa = df[df['location'].str.contains('United States')]
df_usa # just the USA
```

```python
df_usa = df_usa.filter(['location','date','total_cases','new_cases','total_deaths','new_deaths'])
df_usa = df_usa.replace(np.nan,0) # only show data we want, replace NaN's with 0 
df_usa
```
