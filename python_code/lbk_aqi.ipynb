```python
import opendatasets as od
import pandas as pd
import numpy as np

# going to make our full data into a function
def full_data():
    dat=od.download("https://www.kaggle.com/threnjen/40-years-of-air-quality-index-from-the-epa-yearly")
    df = pd.read_csv('40-years-of-air-quality-index-from-the-epa-yearly/aqi_yearly_1980_to_2021.csv', engine = 'python', sep='\s*,\s*')
    #df = pd.read_csv('aqi_daily_1980_to_2021.csv', engine = 'python', sep='\s*,\s*')

    return df
#df # full data set

df = full_data();
df
```

```python
df_lbk = df[df['County'].str.contains('Lubbock')]
df_lbk = df_lbk.sort_values('Year', ascending = False)
df_lbk
```
